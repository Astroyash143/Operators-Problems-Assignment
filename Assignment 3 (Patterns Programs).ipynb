```python
# 16.] Count the total number of digits in a given number
num = int(input("Enter a number: "))

num = abs(num)   
count = 0

if num == 0:
    count = 1
else:
    while num > 0:
        count += 1
        num //= 10

print("Total number of digits:", count)
```

```python
# 17.] Accept any number from the user and check whether it is a 
# perfect number or not

num = int(input("Enter a number: "))

sum_div = 0

for i in range(1, num):
    if num % i == 0:
        sum_div += i

if sum_div == num:
    print(num, "is a Perfect Number")
else:
    print(num, "is NOT a Perfect Number")
```

```python
# 18.] Write a program to calculate the sum of series upto n terms 
# for a given number. Eg. If number is 2 and n_terms=5 the series will be 
# 2+22+222+2222+22222=24690

num = int(input("Enter the number: "))
n = int(input("Enter number of terms: "))

term = 0
sum_series = 0

for i in range(n):
    term = term * 10 + num
    sum_series += term

print("Sum of the series:", sum_series)
```

```python
# 19.] Print all perfect numbers between 1-500

print("Perfect numbers between 1 and 500:")

for num in range(1, 501):
    sum_div = 0

    for i in range(1, num):
        if num % i == 0:
            sum_div += i

    if sum_div == num:
        print(num)
```

```python
# 20.] Print the Fibonacci series for the first 15 terms

a = 0
b = 1
print("Output:")
for i in range(15):
    print(a,end=" ")
    c = a + b
    a = b
    b = c
```
