6.] Accept any year from the user and check whether 
it is a leap year or not.(A leap year is a year which is 
divisible by 4 but not by 100 and should be 
divisible by 400).

```python
year = int(input("Enter the year:"))
if year%4==0 or year%400==0:
    print(f"{year} is a leap year")
elif year%100==0:
    print(f"{year} is not a leap year")
else:
    print(f"{year} is not a leap year")
```

8.] Accept inner and outer radius of a ring and print 
the perimeter and area of the ring 
(Hint: perimeter = 2 π (a+b) , area = π (a*a-b*b) 
I am assuming the radius unit as metre

```python
pi = 3.1428
a = float(input("Enter the outer ring radius in m")) # outer radius must greater than inner radius
b = float(input("Enter the inner ring radius in m")) # inner radius must less than outer radius
perimeter = 2 * pi * (a+b)
area = pi * (a*a-b*b)

print(f"""The perimeter of the ring is {round(perimeter,2)}
and area of the ring is {round(area,2)}""")
```

10.] Consider a room having one door and two windows 
both of the same size. Accept dimensions of the room, 
door and window. Print the area to be painted 
(interior walls) and area to be whitewashed (roof).
Assume the dimensions are in metre


```python

length = float(input("Enter length of the room (m): "))
breadth = float(input("Enter breadth of the room (m): "))
height = float(input("Enter height of the room (m): "))


door_length = float(input("Enter door length (m): "))
door_height = float(input("Enter door height (m): "))


window_length = float(input("Enter window length (m): "))
window_height = float(input("Enter window height (m): "))


wall_area = 2 * (length + breadth) * height
door_area = door_length * door_height
window_area = 2 * window_length * window_height

paint_area = wall_area - door_area - window_area
whitewash_area = length * breadth


print("\nArea to be painted (interior walls):", paint_area, "sq.m")
print("Area to be whitewashed (roof):", whitewash_area, "sq.m")
```

16.] A timestamp is three numbers: a number of hours, 
minutes and seconds. Given two timestamps, 
calculate how many seconds is between them. 
The moment of the first timestamp occurred before the 
moment of the second timestamp.

```python
h1 = int(input("Enter hours of first timestamp: "))
m1 = int(input("Enter minutes of first timestamp: "))
s1 = int(input("Enter seconds of first timestamp: "))

h2 = int(input("Enter hours of second timestamp: "))
m2 = int(input("Enter minutes of second timestamp: "))
s2 = int(input("Enter seconds of second timestamp: "))


time1 = h1 * 3600 + m1 * 60 + s1
time2 = h2 * 3600 + m2 * 60 + s2


difference = time2 - time1

print("Seconds between timestamps:", difference)
```

22.] A cake costs A rupees and B paise. Determine, 
how many rupees and paise cents should one pay for 
N cakes. A program gets three numbers: A, B, N. 
It should print two numbers: total cost in rupees and 
paise. 

```python
A = int(input("Enter rupees (A): "))
B = int(input("Enter paise (B): "))
N = int(input("Enter number of cakes (N): "))

total_paise = (A * 100 + B) * N

total_rupees = total_paise // 100
remaining_paise = total_paise % 100

print("Total cost: ")
print(total_rupees,"rupees", "and", remaining_paise, "paise")
```
