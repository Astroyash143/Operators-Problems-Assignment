### 1.] Write a Python program to replace the last value of tuples in a list. Sample list: [(10, 20, 40), (40, 50, 60), (70, 80, 90)] Expected Output: [(10, 20, 100), (40, 50, 100), (70, 80, 100)]

```python
lst = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]
result = [t[:-1] + (100,) for t in lst]
print(lst)
print(result)
```

### 2.] Write a Python program to remove an empty tuple(s) from a list of tuples.
### Sample data: [(), (), ('',), ('a', 'b'), ('a', 'b', 'c'), ('d')]
### Expected output: [('',), ('a', 'b'), ('a', 'b', 'c'), 'd']


```python
tup = [(), (), ('',), ('a', 'b'), ('a', 'b', 'c'), ('d')]
res = [item for item in tup if item != ()]
print(res)
```

### 4.] Write a Python program to calculate the product, multiplying all the 
### numbers in a given tuple.
### Original Tuple: (4, 3, 2, 2, -1, 18)
### Product - multiplying all the numbers of the said tuple: -864
### Original Tuple: (2, 4, 8, 8, 3, 2, 9)
### Product - multiplying all the numbers of the said tuple: 27648

```python
print("Original tuple: (4, 3, 2, 2, -1, 18)")
tup = (4, 3, 2, 2, -1, 18)
mul = 1
for i in tup:
    mul *= i
print("Product: ",mul)
print()
print("Original tuple: (2, 4, 8, 8, 3, 2, 9)")
mul2 = 1
tup2 = (2, 4, 8, 8, 3, 2, 9)
for i in tup2:
    mul2 *= i
print("Product: ",mul2)
```

### 6.] Write a Python program to convert a tuple of string values to a 
### tuple of integer values.
### Original tuple values: (('333', '33'), ('1416', '55'))
### New tuple values: ((333, 33), (1416, 55))

```python
t = (('333', '33'), ('1416', '55'))
result = tuple(tuple(int(x) for x in inner) for inner in t)
print("Original tuple: ",t)
print("New Tuple: ",result)
```

### 8.] Write a Python program to check if a specified element appears in a tuple of tuples.
### Original list:
### (('Red', 'White', 'Blue'), ('Green', 'Pink', 'Purple'), ('Orange', 'Yellow', 'Lime'))
### Check if White presenet in said tuple of tuples!: True
### Check if White presenet in said tuple of tuples!: True
### Check if Olive presenet in said tuple of tuples!: False

```python
# Original tuple of tuples
colors = (('Red', 'White', 'Blue'),('Green', 'Pink', 'Purple'),('Orange', 'Yellow', 'Lime'))

found = False
for inner in colors:
    if 'White' in inner:
        found = True
        break
print("Check if White present in said tuple of tuples!: ", found)

found = False
for inner in colors:
    if 'Yellow' in inner:
        found = True
        break

print("Check if Yellow present in said tuple of tuples!: ", found)

found = False
for inner in colors:
    if 'Olive' in inner:
        found = True
        break

print("Check if Olive present in said tuple of tuples!: ", found)
```

### 10. Write a Python program to compute the sum of all the elements of each tuple stored inside a list of tuples.
### Original list of tuples: [(1, 2), (2, 3), (3, 4)]
### Sum of all the elements of each tuple stored inside the said list of tuples: [3, 5, 7]
### Original list of tuples: [(1, 2, 6), (2, 3, -6), (3, 4), (2, 2, 2, 2)]
### Sum of all the elements of each tuple stored inside the said list of tuples:[9, -1, 7, 8]

```python
list1 = [(1, 2), (2, 3), (3, 4)]

result1 = []
for t in list1:
    total = 0
    for num in t:
        total += num
    result1.append(total)

print("Original list of tuples:")
print(list1)
print("Sum of all the elements of each tuple stored inside the said list of tuples:")
print(result1)

list2 = [(1, 2, 6), (2, 3, -6), (3, 4), (2, 2, 2, 2)]
result2 = []
for t in list2:
    total = 0
    for num in t:
        total += num
    result2.append(total)

print("\nOriginal list of tuples:")
print(list2)
print("Sum of all the elements of each tuple stored inside the said list of tuples:")
print(result2)
```

### 11. Write a Python program to convert a given list of tuples to a list of lists.
### Original list of tuples: [(1, 2), (2, 3), (3, 4)] 
### Convert the said list of tuples to a list of lists: [[1, 2], [2, 3], [3, 4]]
### Original list of tuples: [(1, 2), (2, 3, 5), (3, 4), (2, 3, 4, 2)]
### Convert the said list of tuples to a list of lists: [[1, 2], [2, 3, 5], [3, 4], [2, 3, 4, 2]]

```python
list1 = [(1, 2), (2, 3), (3, 4)]

result1 = []
for i in list1:
    result1.append(list(i))

print("Original list of tuples:", list1)
print("Convert the said list of tuples to a list of lists:", result1)

list2 = [(1, 2), (2, 3, 5), (3, 4), (2, 3, 4, 2)]

result2 = []
for j in list2:
    result2.append(list(j))

print("\nOriginal list of tuples:", list2)
print("Convert the said list of tuples to a list of lists:", result2)
```
