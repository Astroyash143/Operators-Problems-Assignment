### 1.] Convert a list with duplicates into a set and print all unique values.

```python
data = [5, 'apple', 10, 'banana', 'apple', 5, 'orange', 'banana', 10]
uni_items = set(data)
print(uni_items)
```

### 2.] Given two sets of numbers, find the common elements (intersection).

```python
A = {12,45,212,46,32,78,93,53,17}
B = {43,64,89,23,68,21,68,24,45,46,93,17,93,32}
common = A.intersection(B)
print(common)
```

### 3.] Given two sets, find the elements that are only in the first set.

### 4.] Given two sets, find the elements that are in either set but not both (symmetric difference).

```python
A = {12,45,212,46,32,78,93,53,17}
B = {43,64,89,23,68,21,68,24,45,46,93,17,93,32}
sym_dif = A.symmetric_difference(B)
print(sym_dif)
```

### 10.] Given a set of numbers, check if a number exists inside the set (membership test).

```python
C = {43,64,89,23,68,21,68,24,45,46,93,17,93}
c = int(input("Enter an element to check whether number exists or not:")) 
if c in C:
    stat = True
else:
    stat = False

if stat:
    print(f"-----{c} exists in given set-----")
else:
    print(f"-----{c} does not exist in given set-----")
```

### 16.] Create a set from user input numbers and print the largest number.

```python
num = int(input("Enter the number of elements:"))
a = set({})
i = 1
while num>=i:
    inp = int(input("Enter the element"))
    a.add(inp)
    i+=1
print("Maximum element from the input set is :",max(a))
```
