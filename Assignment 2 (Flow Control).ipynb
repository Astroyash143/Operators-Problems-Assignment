```python
# 1.] Write a python program to read three numbers and if any two 
# variables are equal, print that number

a = int(input("Enter first number: (a) "))
b = int(input("Enter Second number: (b) "))
c = int(input("Enter third number: (c) "))

if a==b==c:
    print(f"a,b and c all numbers are equal or same")
elif a==b and (b!=c or a!=c):
    print(f"a and b are equal and c is different")
elif a==c and (a!=b or c!=b):
    print(f"a and c are equal and c is different")
elif b==c and (b!=a or c!=a):
    print(f"b and c are equal and a is different")
else:
    print(f"a,b and c all numbers are not equal or different")
```

```python
# 2.] Write a python program to read two numbers and find the 
# smallest among them using ternary operator
# A ternary operator is a single-line conditional expression that selects one of two values based on a condition and stores it in a variable.

a = int(input("Enter first number: (a) "))
b = int(input("Enter Second number: (b) "))

max = a if a>b else b
print(f"{max} is the maximum number compared between {a} and {b}")
```

```python
# 3.] Write a python program to read three numbers and find the 
# largest among them without ternary operator

a = int(input("Enter first number: (a) "))
b = int(input("Enter Second number: (b) "))
c = int(input("Enter third number: (c) "))

max = a if a>b and b>c else c
print(f"{max} is the maximum number compared between {a},{b} and {c}")
```

```python
# 5.] Write a python program to read four numbers (representing the 
# four octets of an IP) and check whether they all are in the range 
# between 0 to 255 
# eg: 255.255.255.255
# Check all octets in 0-255 range


a = int(input("Enter first number: (a) "))
b = int(input("Enter Second number: (b) "))
c = int(input("Enter third number: (c) "))
d = int(input("Enter fourth number: (a) "))

print()
print()

if 0<=a<=255 and 0<=b<=255 and 0<=c<=255 and 0<=d<=255:
    print(f"{a}.{b}.{c}.{d} is valid IP number")
else:
    print(f"{a}.{b}.{c}.{d} is not valid IP number")
```

```python
# 7.] Write a python program to read a number ,if it is an even number , 
# print the square of that number and if it is odd number print cube of 
# that number.

number = int(input("Enter the number: "))

if number%2==0:
    print(f"{number} is even and its Square is {number**2}")
else:
    print(f"{number} is odd and its Cube is {number**3}")
```

```python
# 9.] Write a Python program to find the sum of all even numbers from 
# 0 to 10
sum = 0
for i in range(0,11):
    if i%2==0:
        sum+=i
    else:
        continue
print("Sum of all even values from 0 to 10 is",sum)
```

```python
# 10.] Write a python program to read three numbers (a,b,c) and check 
# how many numbers between ‘a’ and ‘b’ are divisible by ‘c’

a = int(input("Enter first number: (a) "))
b = int(input("Enter Second number: (b) "))
c = int(input("Enter third number: (c) "))


if a%c==0 and b%c==0:
    print(f"{a} and {b} both numbers are divisible by {c}")
elif a%c==0 and b%c!=0: 
    print(f"Only {a} is divisible by {c}")
elif a%c!=0 and b%c==0:
    print(f"Only {b} is divisible by {c}")
else: 
    print(f"{a} and {b} both numbers are not divisible by {c}")
```

```python
# 12.] Write a python program to print the factorial of a given number

num = int(input("Enter the number whose factorial has to find: "))
fact=1
for i in range(1,num+1):
    fact*=i
print(f"Factorial of the number {num}! is {fact}")
```

```python
# 13.] Write a python program to print the first 18 terms of a Fibonacci 
# series

a = 0
b = 1

print("First 18 Fibonacci numbers:")

for i in range(18):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```
