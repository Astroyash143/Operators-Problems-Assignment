## 1.] Unary Operator 
<br>
Q1. Accept an integer from the user and:
Convert it to its negative using a unary operator
Convert it back to positive using a unary operator
Print both results

```python
number = int(input("Enter an integer: "))
number = -number
print("Number after converting it to negative using unary operator:", number)
number = -number
print("Number after converting it to Positive using unary operator:", number)
```

## 5.] Bitwise Operator
<br>
Q9. Accept two integers from the user and:
Perform bitwise AND, OR, XOR, Left Shift, Right Shift, One's Complement
Print the result of each operation

```python
a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))

print("Bitwise AND (a & b):", a & b)
print("Bitwise OR  (a | b):", a | b)
print("Bitwise XOR (a ^ b):", a ^ b)

print("Left Shift  (a << 1):", a << 1)
print("Right Shift (a >> 1):", a >> 1)

print("One's Complement of a (~a):", ~a)
print("One's Complement of b (~b):", ~b)
```
