## 1.] Remove all duplicate items from a list but keep the original order. (Used in cleaning API/CSV data)

```python
lst = [1,5,3,3,7,2,6,3,6]
unique = []
seen = set()

for i in lst:
    if i not in seen:
        seen.add(i)
        unique.append(i)

print(unique)

```

## 5.] Find the most frequent element in a list. (Used in analytics & dashboards)

```python
L = [1,4,6,4,6,5,7,3,5,4,6,7]
flist = []
for i in L:
    flist.append(L.count(i))

mfrq = max(flist)
idx = flist.index(mfrq)

print(f"""{L[idx]} is the number with highest frequent 
among all other items and its frequency is {mfrq}""")
```

## 16.] Given a list of transactions (credit + / debit -), return total balance, total credit, and total debit.

```python
trs = [-1000,3700,400,-480,300,520,-340]
cred = []
deb = []
bal = 0
tc = 0
td = 0
for i in trs:
    bal += i
    if i<0:
        deb.append(i)
    else:
        cred.append(i)
for j in cred:
    tc += j

for k in deb:
    td += k

print(f"Total Balance: {bal}")
print(f"Total Credit: {tc}")
print(f"total Debit: {td}")
```

## 20.] Given a sorted list,insert a new element at correct position (binary insertion).

```python
lst = [2,4,5,8,10,16,20]
inp = int(input("Enter the element to be inserted:"))
low = 0
high = len(lst)

while low < high:
    mid = (low + high) // 2
    if lst[mid] < inp:
        low = mid + 1
    else:
        high = mid
        
lst.insert(low, inp)
print(lst)
```

## 23.] Given a list of API response times in ms, find the 95th percentile manually.

```python
api_response_times_ms = [120, 245, 98, 310, 450, 180, 220, 95, 600, 150]
sorted_data = sorted(api_response_times_ms)
n = len(sorted_data)
position = 0.95 * (n + 1)

if position >= n:
    p95 = sorted_data[-1]
else:
    lower_index = int(position) - 1
    upper_index = lower_index + 1
    fraction = position - int(position)
    p95 = sorted_data[lower_index] + fraction * (
        sorted_data[upper_index] - sorted_data[lower_index]
    )

print("Sorted data:", sorted_data)
print("95th percentile:", p95, "ms")
```

## 24.] Given student marks, curve the marks (increase each mark until class avg = 70).

```python
marks = [56,35,65,45,54,23,58,35,70]

current_avg = sum(marks) / len(marks)
increment = 70 - current_avg
curved_marks = [round(mark + increment, 2) for mark in marks]
new_avg = sum(curved_marks) / len(curved_marks)
print("Original marks: ", marks)
print("Current average: ", current_avg)
print("Increment added to each mark: ", increment)
print("Curved marks: ", curved_marks)
print("New average: ", new_avg)
```
