```python
# 5.] Write a Python program to print the number of prime numbers 
# which are less than or equal to an given integer. 
# Input: Input the number(n):35
# Number of prime numbers which are less than or equal to n.: 11

inp = int(input("Input the number:"))
print()
cnt = 0
for num in range(2, inp):   # 2 to 35
    
    is_prime = True
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")

    if is_prime==True:
        cnt+=1
print("\nNumber of prime numbers which are less than or equal to n.:",cnt)
```

```python
# # 7.] Write a Python program to find the longest common prefix 
# # string amongst an given array of strings. Return false If there is no 
# # common prefix.
# # For Example, longest common prefix of "abcdefgh" and "abcefgh" 
# # is "abc". 
# # Sample Output : abc

n = int(input("Enter the number of strings: "))
arr = []
for i in range(n):
    arr.append(str(input(f"Enter the {i} string: ")))


def longest_common_prefix(arr):
    prefix = arr[0]

    for s in arr[1:]:
        while not s.startswith(prefix):
            arr = prefix[:-1]
            if prefix == "":
                return False
    return prefix

longest_common_prefix(arr)
```

```python
# 17.] Arrange String characters such that lowercase letters should come 
# first
# Given input String of combination of the lower and upper case 
# arrange characters in such a way that all lowercase letters should come 
# first.

ips = input("Enter the string with no spaces and Special characters: ")
lst = list(ips)
l = []
u = []
for i in lst:
    if i.islower()!=True:
        u.append(i)
    else:
        l.append(i)

newlst = l + u
opstr = "".join(newlst)
print(opstr)  
```

```python
# 18.] Given a string input Count all lower case, upper case, digits, 
# and special symbols
# eg :    Total counts of chars, digits,and symbols Chars = 8 Digits = 3 
# Symbol = 4
countL = 0   # counter for Lowercase
countU = 0   # counter for Uppercase.
countN = 0   # counter for Digits
countS = 0   # counter for Special Symbols

ipstr = input("Enter the String: ")

for i in ipstr:
    if i.islower():
        countL += 1
    elif i.isupper():
        countU += 1
    elif i.isdigit():
        countN += 1
    else:
        countS += 1

print("Count of Lowercase letters: ",countL)
print("Count of Uppercase letters: ",countU)
print("Count of Digit letters: ",countN)
print("Count of Specialcase letters: ",countS)
    
```

```python
# 22.] Given a string, return the sum and average of the digits that appear 
# in the string, ignoring all other characters
# eg :     input = "English = 78 Science = 83 Math = 68 History = 65"
# output =  sum 294 Percentage is 73.5

s = input("Enter the string:")

numbers = []
num = ""

for ch in s:
    if ch.isdigit():
        num += ch                      # here the input string is traversed in for loop and when digit comes it stores in a string called num.                                                  
    else:                               # and when number ends that theres nothing or "" after the number completition it stores                                    
        if num != "":                   # the whole number as integer... it happens till end of the string
            numbers.append(int(num))
            num = ""

total = sum(numbers)
average = total / len(numbers)

print("Sum of numbers =", total)
print("Percentage of numbers is", average)
```
